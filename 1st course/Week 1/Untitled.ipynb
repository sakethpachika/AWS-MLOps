```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/sakethpachika/Desktop/Book1.csv")
```

```python
def to_sentiment(star_rating):
    if star_rating in {1, 2}: # negative
        return -1 
    if star_rating == 3:      # neutral
        return 0
    if star_rating in {4, 5}: # positive
        return 1
```

```python
df['sentiment'] = df['Rating'].apply(lambda star_rating: 
    to_sentiment(star_rating) 
)
```
